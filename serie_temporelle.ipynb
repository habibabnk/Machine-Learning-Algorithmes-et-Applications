```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv("C:\\Users\\ThikPad\\Downloads\\monthly_sales.csv")  
```

```python

df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace=True)
df.head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df['Sales'],marker='.')
plt.title('Ventes mensuelles')
plt.xlabel('Mois')
plt.ylabel('Ventes')
plt.show()
```
